## subplots


```python
x=np.linspace(1,15,10)
y=x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(x,y,'r*')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y)

ax2.plot(y,x)
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
import matplotlib.pyplot as plt
labels=["php","python","java","c"]
numbers=[75,70,80,60]
fig=plt.figure(figsize=(10,10))
axes1=fig.add_axes([0.1,0.1,1,1])
axes1.pie(numbers,labels=labels)
axes1.pie(numbers,labels=labels,autopct="%0.2f%%")
```

```python
import matplotlib.pyplot as plt
labels=["php","python","java","c"]
numbers=[75,70,80,60]
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,1,1])

axes1.pie(numbers,labels=labels,autopct="%0.2f%%",colors=["orange","pink","green","blue"])
```

```python
Explode=[0.1,0.7,0,0]
plt.pie(numbers,labels=labels,explode=Explode)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
df.info()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

```python
sns.scatterplot(x="size",y="total_bill",data=df)
```

```python
sns.lineplot(x="size",y="total_bill",data=df)
```

```python
sns.distplot(df["tip"]) #most of the tips range from 2-4.
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="day")
```

```python
sns.barplot(data=df,x="day",y="tip")
```

```python
sns.barplot(data=df,x="sex",y="tip",hue="day")
```

```python
sns.countplot(x="sex",data=df)
```

```python
sns.jointplot(x="total_bill",y="tip",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
#1st line - max value
#2nd line upper quartile
#3rd median
#4th lower quartile
#5th min value
```

```python
corr=df.corr()
corr
```
