```python
df=sns.load_dataset('car_crashes')
car_crashes = sns.load_dataset('car_crashes')
```

```python
car_crashes['percent_speeding'] = (car_crashes['speeding']/car_crashes['total'])*100
car_crashes['percent_alcohol'] = (car_crashes['alcohol']/car_crashes['total'])*100
car_crashes['percent_not_distracted'] = (car_crashes['not_distracted']/car_crashes['total'])*100
car_crashes['percent_no_previous'] = (car_crashes['no_previous']/car_crashes['total'])*100


car_crashes.head()
```

```python
data = car_crashes.sort_values(by='total', ascending=False)
print(data[['abbrev','total','alcohol']].head())
fig, ax = plt.subplots(figsize=(10,15))
ax0=sns.barplot(data=data,x='total',y='abbrev')
ax0=sns.barplot(data=data,x='alcohol',y='abbrev', color='grey')
ax0.set_title('Barplot of Total Collisionsn(Grey bar indicates number of people intoxicated)')
ax0.set(xlabel='Number of Drivers in Fatal Collisions', ylabel='State');
```

##### South Carolina, North Dakota and West Virginia have approximately 24 people involed in fatal collisions. Approximately 10 individauls out of 24 in South Carolina and North Dakota had caused alcohol related fatal accidents.

```python
data_alcohol = car_crashes.sort_values(by='alcohol', ascending=False)
data_alcohol = data_alcohol.head(10)

data_percent_alcohol = car_crashes.sort_values(by='percent_alcohol', ascending=False)
data_percent_alcohol = data_percent_alcohol.head(10)


data_speeding = car_crashes.sort_values(by='speeding', ascending=False)
data_speeding = data_speeding.head(10)

data_percent_speeding = car_crashes.sort_values(by='percent_speeding', ascending=False)
data_percent_speeding = data_percent_speeding.head(10)
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5), squeeze=False)
ax00 = sns.barplot(data=data_alcohol,x='alcohol',y='abbrev', ax=ax[0,0])
ax00.set_title('Top 10 Alcohol Related Collisions')
ax01 = sns.barplot(data=data_percent_alcohol,x='percent_alcohol',y='abbrev', ax=ax[0,1])
ax01.set_title('Top 10 Portion of Alcohol Related Collisions');
```

#### Even though North Dakota has the highest number of Alcohol related fatalities, it is Montana that has the highest porportion of Alcohol related fatalities.

```python
fig, ax = plt.subplots(1,2,figsize=(10,5), squeeze=False)
ax00 = sns.barplot(data=data_speeding,x='speeding',y='abbrev', ax=ax[0,0])
ax00.set_title('Top 10 States with Speeding \n Related Collisions')
ax01 = sns.barplot(data=data_percent_speeding,x='percent_speeding',y='abbrev', ax=ax[0,1])
ax01.set_title('Top 10 States with Portion of \n Speeding Related Collisions');
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5), squeeze=False)
ax00 = sns.barplot(data=data_speeding,x='speeding',y='abbrev', ax=ax[0,0])
ax00.set_title('Top 10 States with Speeding \n Related Collisions')
ax01 = sns.barplot(data=data_percent_speeding,x='percent_speeding',y='abbrev', ax=ax[0,1])
ax01.set_title('Top 10 States with Portion of \n Speeding Related Collisions');
```
